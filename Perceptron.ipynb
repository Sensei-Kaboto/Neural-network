```python
import numpy as np
import matplotlib.pyplot as plt

lr = 0.003
n_epochs = 10
```

## Generate Data Randomly

```python
input = np.random.randint(10,size=(10,2))
output =np.random.randint(2, size =(10,1))
weight = np.round(np.random.uniform(-1,1, size = (input.shape[0],input.shape[1])),2)
print(f"the input : {input}")
print(f"the output : {output}")
print(f"the weight vector : {weight}")
```

## Training

```python

# the activation function / sigmoid function
def sigmoid(z):
  if z >= 0:
    return 1
  else :
    return 0
#the training function
def guess(input):
  bias = np.random.rand(1)
  sum = 0
  output =np.zeros(input.shape[0],dtype=int)
  for i in range(len(input)):
    for j in range(weight.shape[1]):
      sum += input[i,j] * weight[i,j]
    output[i] =sigmoid(sum + bias)
  return output

def train(input,target,weight):
      g = guess(input)
      error =np.ones(len(g), dtype=int)
      print(f"the guess output : {g}")
      for i in range(len(g)):
        error[i] = target[i] - g[i]
      print(f"the error : {error}")
      print("--------------------------------------------------------------------------")
      # tune all the weights
      product = input * lr
      for i in range(len(weight)):
        for j in range(weight.shape[1]):
          weight[i,j] += error[i] *  product[i,j]
      return g
     

print(weight)
for i in range(n_epochs):
  print(output.T)
  y_pred = train(input,output,weight)
print(weight)
```
